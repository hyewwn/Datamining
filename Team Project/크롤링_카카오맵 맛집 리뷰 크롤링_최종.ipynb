```python
# 서울맛집리스트 불러오기
seoul = pd.read_csv('서울맛집리스트_구이름.csv')
seoul.head()
```

```python
from time import sleep

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import StaleElementReferenceException
from bs4 import BeautifulSoup

options = webdriver.ChromeOptions()
options.add_argument('headless')
options.add_argument('lang=ko_KR')
chromedriver_path = "chromedriver"
driver = webdriver.Chrome(os.path.join(os.getcwd(), chromedriver_path), options=options)  # chromedriver 열기
```

```python
def main():
    global driver, load_wb, review_num, fin_result
    
    fin_result = []

    driver.implicitly_wait(4)  # 렌더링 될때까지 4초 기다리기
    driver.get('https://map.kakao.com/')  # 주소 가져오기
    
    # 검색할 목록
    for i in range(len(seoul)) :
        print(datetime.now().strftime('%H:%M:%S'))
        print('=====', i, '번째 맛집 : ', seoul.iloc[i][0], '=====')
        a = search(seoul.iloc[i][0])
        print('search 결과=========\n', pd.DataFrame(a, columns = ['이름', '위치', '리뷰', '리뷰일자', '평점']).head(5))
        if a is None :
            print('여길 들어왔다.')
            continue
        print('여길 나왔다.')
        for i in range(len(a)) :
            print('얼마 남지 않았다!!', end = ' ')
            fin_result.append(a[i])
        print('\n여기에 있다.\n', fin_result[-20 : ], '\n')

    print('fin_result==========\n', fin_result[-20 : ])
    driver.quit()
    print("finish")
    
    time = datetime.now().strftime('%m_%d__%H_%M_%S')
    pd.DataFrame(fin_result,
                 columns = ['이름', '위치', '리뷰', '리뷰일자', '평점']).to_csv('카카오맵_{}.csv'.format(time), index = False, encoding = 'utf-8-sig')

def search(place):
    result_list = []
    global driver

    search_area = driver.find_element_by_xpath('//*[@id="search.keyword.query"]')  # 검색 창
    search_area.send_keys(place)  # 검색어 입력
    driver.find_element_by_xpath('//*[@id="search.keyword.submit"]').send_keys(Keys.ENTER)  # Enter로 검색
    sleep(3)

    # 검색된 정보가 있는 경우에만 탐색
    # 1번 페이지 place list 읽기
    html = driver.page_source

    soup = BeautifulSoup(html, 'html.parser')
    place_lists = soup.select('.placelist > .PlaceItem') # 검색된 장소 목록

    # 검색된 첫 페이지 장소 목록 크롤링하기
    a = crawling(place, place_lists)
    print(a)
    print('크롤링은 함')
    for i in range(len(a)) :
        result_list.append(a[i])

    driver.implicitly_wait(20)
    search_area.clear()
    driver.implicitly_wait(20)
    
    if a == [] :
        # 우선 더보기 클릭해서 2페이지
        try:
            driver.find_element_by_xpath('//*[@id="info.search.place.more"]').send_keys(Keys.ENTER)
            sleep(3)
            
            # 2~ 5페이지 읽기
            for i in range(2, 6):
                # 페이지 넘기기
                xPath = '//*[@id="info.search.page.no' + str(i) + '"]'
                driver.find_element_by_xpath(xPath).send_keys(Keys.ENTER)
                sleep(3)
                
                html = driver.page_source
                soup = BeautifulSoup(html, 'html.parser')
                place_lists = soup.select('.placelist > .PlaceItem') # 장소 목록 list
                
                b = crawling(place, place_lists)
                for i in range(len(b)) :
                    result_list.append(b[i])
                if b != [] :
                    break
        
        except ElementNotInteractableException:
            print('not found')
        finally:
            search_area.clear()
    
    return result_list


def crawling(place, place_lists): # 페이지 목록을 받아서 크롤링 하는 함수, :param place: 리뷰 정보 찾을 장소이름
    global driver
    while_flag = False
    result = []
    target_place = place[3 :] # 내 검색어
    if place_lists == [] :
        return result
    
    for i, place in enumerate(place_lists):
        # place : 검색 결과
        place_name = place.select('.head_item > .tit_name > .link_name')[0].text  # place name
        if target_place != place_name.replace(' ', '') :
            print('다른 거 등장!', place_name)
            if place_name.replace(' ', '') in target_place :
                print('1')
                pass
            elif place_name.replace('점','').replace(' ', '') in target_place.replace('점', '') :
                print('2')
                pass
            elif place_name.replace('본점', '').replace(' ', '') in target_place.replace('본점', '') :
                print('3')
                pass
            elif place_name.replace('점', '').replace(' ', '') in target_place.replace('본점', '') :
                print('4')
                pass
            elif place_name.replace('본점', '').replace(' ', '') in target_place.replace('점', '') :
                print('5')
                pass
            elif place_name.replace('본점', '').replace(' ', '') in target_place :
                print('6')
                pass
            elif place_name.replace('점', '').replace(' ', '') in target_place :
                print('7')
                pass
            elif place_name.replace(' ', '') in target_place.replace('점', '') :
                print('8')
                pass
            elif place_name.replace(' ', '') in target_place.replace('본점', '') :
                print('9')
                pass
            elif target_place in place_name.replace(' ', '') :
                print('10')
                pass
            
            else :
                continue
                        
        place_address = place.select('.info_item > .addr > p')[0].text  # place address
        
        #print(driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[' + str(i + 1) + ']').text)
        
        detail_page_xpath = '//*[@id="info.search.place.list"]/li[' + str(i + 1) + ']/div[5]/div[4]/a[1]'
        driver.find_element_by_xpath(detail_page_xpath).send_keys(Keys.ENTER)
        driver.switch_to.window(driver.window_handles[-1])  # 상세정보 탭으로 변환
        sleep(3)
        
        print('####', place_name)
        
        #try :
        #    if driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[' + str(6) + ']') is not None :
        
        ##if driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[' + str(6) + ']') is not None :
         #       detail_page_xpath = '//*[@id="info.search.place.list"]/li[' + str(i + 2) + ']/div[5]/div[4]/a[1]'
         #       driver.find_element_by_xpath(detail_page_xpath).send_keys(Keys.ENTER)
         #       print('야@')
         #       driver.switch_to.window(driver.window_handles[-1])  # 상세정보 탭으로 변환
         #       sleep(3)
        
          #      print('####', place_name)
        ##else :
        #except NoSuchElementException :
        
        ##if place.select('.list > li.AdItem')
        ##if driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[' + str(i + 1) + ']').text != place_name
        
        ##//*[@id="info.search.place.list"]/li[4]
        ##info\.search\.place\.list > li.AdItem
        ##if driver.find_element_by_xpath('//*[@id="info.search.place.list"]/li[' + str(i + 1) + ']').class == 'AdItem' :

          #  detail_page_xpath = '//*[@id="info.search.place.list"]/li[' + str(i + 1) + ']/div[5]/div[4]/a[1]'
          #  driver.find_element_by_xpath(detail_page_xpath).send_keys(Keys.ENTER)
          #  driver.switch_to.window(driver.window_handles[-1])  # 상세정보 탭으로 변환
          #  sleep(3)
        
         #   print('####', place_name)

        # 첫 페이지
        a = extract_review(place, place_name)
        for i in range(len(a)) :
            result.append(a[i])
        print(len(a), 'a')
        
        # 2-5 페이지
        idx = 3
        try:
            page_num = len(driver.find_elements_by_class_name('link_page')) # 페이지 수 찾기
            for i in range(page_num-1):
                # css selector를 이용해 페이지 버튼 누르기
                driver.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_review > div > a:nth-child(' + str(idx) +')').send_keys(Keys.ENTER)
                sleep(3)
                b = extract_review(place, place_name)
                for i in range(len(b)) :
                    result.append(b[i])
                print(len(b), 'b')
                idx += 1
            driver.find_element_by_link_text('다음').send_keys(Keys.ENTER) # 5페이지가 넘는 경우 다음 버튼 누르기
            sleep(3)
            c = extract_review(place, place_name) # 리뷰 추출
            for i in range(len(c)) :
                result.append(c[i])
            print(len(c), 'c')
        except (NoSuchElementException, ElementNotInteractableException):
            print("no review in crawling")

        # 그 이후 페이지
        while True:
            idx = 4
            try:
                page_num = len(driver.find_elements_by_class_name('link_page'))
                for i in range(page_num-1):
                    driver.find_element_by_css_selector('#mArticle > div.cont_evaluation > div.evaluation_review > div > a:nth-child(' + str(idx) +')').send_keys(Keys.ENTER)
                    sleep(3)
                    d = extract_review(place, place_name)
                    for i in range(len(d)) :
                        result.append(d[i])
                    print(len(d), 'd')
                    idx += 1
                driver.find_element_by_link_text('다음').send_keys(Keys.ENTER) # 10페이지 이상으로 넘어가기 위한 다음 버튼 클릭
                sleep(3)
                e = extract_review(place, place_name) # 리뷰 추출
                for i in range(len(e)) :
                    result.append(e[i])
                print(len(e), 'e')
            except (NoSuchElementException, ElementNotInteractableException):
                print("no review in crawling")
                break

        driver.close()
        sleep(2)
        driver.switch_to.window(driver.window_handles[0]) # 검색 탭으로 전환
        sleep(2)
        return result


def extract_review(place, place_name):
    global driver
    
    list_from_extract = []
    place_name = place.select('.head_item > .tit_name > .link_name')[0].text  # place name
    place_address = place.select('.info_item > .addr > p')[0].text  # place address

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 첫 페이지 리뷰 목록 찾기
    review_lists = soup.select('.list_evaluation > li')

    # 리뷰가 있는 경우
    if len(review_lists) != 0:
        for i, review in enumerate(review_lists):
            comment = review.select('.txt_comment > span') # 리뷰
            comment_time = review.select('.append_item > span.time_write')[0].text
            rating = review.select('.grade_star > em') # 별점
                        
            list_from_extract.append((place_name, place_address, comment[0].text, comment_time, rating[0].text.replace('점','')))
    
    else:
        print('no review in extract')

    return list_from_extract

if __name__ == "__main__":
    main()
```
